# matplotlib visualizations for data science

```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
x = np.linspace(1,10,100)
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))
```

```python
plt.plot(x,np.sin(x),'--')
plt.plot(x,np.cos(x),'-.')
```

```python
fig = plt.figure()
ax = fig.add_axes([0.2,0.2,0.9,0.9])

x= np.arange(0,math.pi*2, 0.05)
y = np.cos(x)

ax.plot(x,y)
ax.set_title("cos wave")
ax.set_xlabel("angle")
ax.set_ylabel("cos value")
```

```python

x = np.linspace(0,2*np.pi,400)
y=np.sin(x**2)+np.cos(x)

fig,ax=plt.subplots()
ax.plot(x,y)
ax.set_title('plot in matplotlib')
```

```python
plt.figure(figsize=(16,12))

fig, ax = plt.subplots(2,sharex='col',sharey='row')

ax[0].plot(x)
ax[1].plot(y)
```

### setting interactive background for matplotlib

```python
import matplotlib
matplotlib.use('nbagg')
%matplotlib inline

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy import stats
import pandas as pd
```

```python
sns.set_palette("BuPu_d",desat=0.6)
sns.set_context("notebook",font_scale=2.0)
```

```python
np.random.seed(42)
x= stats.gamma(5).rvs(420)
y= stats.gamma(13).rvs(420)

with sns.axes_style("white"):
  sns.jointplot(x,y,kind="hex", size=8)
```

```python
from scipy.stats import norm, rayleigh

a = rayleigh.rvs(loc=5, scale=2, size=1000) + 1
b = rayleigh.rvs(loc=5, scale=2, size=1000)
c = rayleigh.rvs(loc=5, scale=2, size=1000) - 1
```

```python
data = pd.DataFrame({"a": a, "b": b, "c": c}, columns=["a", "b", "c"])
```

```python
axes = data.plot(kind="hist", stacked=True, bins=50, figsize=(16, 8),color=['r','b','g'],grid=True)
axes.set_title("Fabricated Wind Speed Data", fontsize=20)
axes.set_xlabel("Mean Hourly Wind Speed (km/hr)", fontsize=16)
_ = axes.set_ylabel("Velocity Counts", fontsize=16)
```
