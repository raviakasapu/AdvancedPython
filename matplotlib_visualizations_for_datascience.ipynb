# matplotlib visualizations for data science

```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
x = np.linspace(1,10,100)
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))
```

```python
plt.plot(x,np.sin(x),'--')
plt.plot(x,np.cos(x),'-.')
```

```python
fig = plt.figure()
ax = fig.add_axes([0.2,0.2,0.9,0.9])

x= np.arange(0,math.pi*2, 0.05)
y = np.cos(x)

ax.plot(x,y)
ax.set_title("cos wave")
ax.set_xlabel("angle")
ax.set_ylabel("cos value")
```

```python

x = np.linspace(0,2*np.pi,400)
y=np.sin(x**2)+np.cos(x)

fig,ax=plt.subplots()
ax.plot(x,y)
ax.set_title('plot in matplotlib')
```

```python
plt.figure(figsize=(16,12))

fig, ax = plt.subplots(2,sharex='col',sharey='row')

ax[0].plot(x)
ax[1].plot(y)
```
